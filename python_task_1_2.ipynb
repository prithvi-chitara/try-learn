```python
# Q8. TAKE A USER INPUT AS INTEGER N. FIND OUT THE SUM FROM 1 TO N. IF ANY NUMBER IF DIVISBLE BY 5, THEN SKIP THAT NUMBER. AND IF THE SUM IS GREATER THAN 300, DO NOT NEED TO CALCULTE THE SUM FURTHER MORE. PRINT THE FINAL RESULT. AND DO NOT USE FOR LOOP TO SOLVE THE PROBLEM.
```

```python
num = int(input("Enter the number: "))
sum = 0
for i in range(1,num+1):
  if i%5==0:
    continue
  sum += i

  if sum>300:
    sum -=i
    break
  i+=1
print(sum)
```

```python
# Q9. WAP tha keeps on accepting number from the user until enters zero.
#display the sum and average of all the numbers.
# sum all number and average
```

```python
sum = 0
count = 0
while True:
  num = int(input("Enter number: "))
  if num==0:
    break
  sum = sum+num
  count = count+1
print("sum", sum)
print("average is ",sum/count)
```
